```python
# q6)Try to identify important or relevant columns for feature extraction

def difference_in_days(melt, lags, ffday, customer_id_col, create_date_col, net_amount_col):
    melt = melt.copy()
    grouped = melt.groupby([customer_id_col])
    
    for i in range(ffday, lags+1):
        melt[f'Last-{i}day_Sales'] = grouped[net_amount_col].shift(i)
        melt[f'Last-{i}day_Diff'] = grouped[net_amount_col].diff(i).fillna(0)  # Use fillna(0) to handle missing values
    
    return melt

# Apply the difference_in_days function to the dataframe
lags = 7
ffday = 1
customer_id_col = 'unique_cust_id'
create_date_col = 'ORDER_CREATION_DATE'
net_amount_col = 'amount_in_usd'

df_transformed = difference_in_days(df, lags, ffday, customer_id_col, create_date_col, net_amount_col)
df_transformed
```

```python
# q1)Modify the dataset to pass into any type of machine learning models. 


import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# Step 1: Handle Missing Values
# Handle missing values in the dataset based on your chosen strategy

# Step 2: Encode Categorical Variables
# Encode categorical variables if any using one-hot encoding or label encoding

# Step 3: Split into Features and Target
X = df.drop('amount_in_usd', axis=1)
y = df['amount_in_usd']

# Exclude datetime columns from scaling
datetime_cols = X.select_dtypes(include=['datetime64']).columns
X_scaled = X.drop(datetime_cols, axis=1)

# Step 4: Split into Training and Testing Sets
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

print("X_train shape:", X_train.shape)
print("y_train shape:", y_train.shape)
print("X_test shape:", X_test.shape)
print("y_test shape:", y_test.shape)

# Step 5: Normalize or Standardize Features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Now you can pass the preprocessed X_train_scaled, y_train, X_test_scaled, y_test
# into any type of machine learning model for training and evaluation
```

```python
# q2 Try different machine learning models like - 
#Linear Regression
#Support Vector Machine 
#Decision Tree
#Random Forest 
#AdaBoost
#Xgboost etc. 


from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, AdaBoostRegressor
import xgboost as xgb
from sklearn.impute import SimpleImputer
from sklearn.metrics import mean_squared_error
import numpy as np

# Initialize the imputer
imputer = SimpleImputer(strategy='mean')

# Impute missing values
X_train_imputed = imputer.fit_transform(X_train_scaled)
X_test_imputed = imputer.transform(X_test_scaled)

# Initialize the models
linear_regression = LinearRegression()
support_vector_machine = SVR()
decision_tree = DecisionTreeRegressor()
random_forest = RandomForestRegressor()
adaboost = AdaBoostRegressor()
xgboost = xgb.XGBRegressor()

# Train and evaluate the models
models = [linear_regression,decision_tree,random_forest,adaboost,xgboost]#   
model_names = ['Linear Regression','Decision Tree','Random Forest','AdaBoost','XGBoost']#   

for model, name in zip(models, model_names):
    # Train the model
    model.fit(X_train_imputed, y_train)

    # Predict on the test set
    y_pred = model.predict(X_test_imputed)

    # Evaluate the model
    mse = mean_squared_error(y_test, y_pred)
    rmse = np.sqrt(mse)

    # Print the evaluation result
    print(f"{name}: RMSE = {rmse}")
```

```python
# q3)Perform Regression model evaluations like MSE, RMSE, R-Square etc.

from sklearn.metrics import mean_squared_error, r2_score

# Train and evaluate the models
for model, name in zip(models, model_names):
    # Train the model
    model.fit(X_train_imputed, y_train)

    # Predict on the test set
    y_pred = model.predict(X_test_imputed)

    # Calculate MSE
    mse = mean_squared_error(y_test, y_pred)

    # Calculate RMSE
    rmse = np.sqrt(mse)

    # Calculate R-Squared
    r2 = r2_score(y_test, y_pred)

    # Print the evaluation results
    print(f"{name}:")
    print(f"MSE = {mse}")
    print(f"RMSE = {rmse}")
    print(f"R^2 = {r2}")
    print()
```

```python
# q4) Perform Regression model evaluations like MSE, RMSE, R-Square etc.

# Initialize empty lists to store evaluation metrics
mse_values = []
rmse_values = []
r2_values = []

# Train and evaluate the models
for model, name in zip(models, model_names):
    # Train the model
    model.fit(X_train_imputed, y_train)

    # Predict on the test set
    y_pred = model.predict(X_test_imputed)

    # Calculate MSE
    mse = mean_squared_error(y_test, y_pred)

    # Calculate RMSE
    rmse = np.sqrt(mse)

    # Calculate R2
    r2 = r2_score(y_test, y_pred)

    # Append the evaluation metrics to the lists
    mse_values.append(mse)
    rmse_values.append(rmse)
    r2_values.append(r2)

# Print the evaluation metrics for each model
for name, mse, rmse, r2 in zip(model_names, mse_values, rmse_values, r2_values):
    print(f"{name}:")
    print(f"MSE = {mse:.4f}")
    print(f"RMSE = {rmse:.4f}")
    print(f"R2 = {r2:.4f}")
    print()

# Find the index of the most efficient model based on RMSE
best_model_index = np.argmin(rmse_values)
best_model_name = model_names[best_model_index]

# Print the most efficient model
print(f"The most efficient model based on RMSE is: {best_model_name}")
```
